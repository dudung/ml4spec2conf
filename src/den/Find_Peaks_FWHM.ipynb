# Finding Peaks and FWHM

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
import scipy.signal as scs
```

```python
# Load dataset
url = "../../data/raw/Exp/Dim_Si_190_5_par.xlsx"
names = ['lam', 'csca'] 
df = pd.read_excel(url, names=names, header=None)

x = df['lam'].values
y = df['csca'].values
```

```python
#interpolate
f_int = interp1d(x, y, kind='cubic')
x_int = np.linspace(min(x), max(x), 1000)
y_int = f_int(x_int)
```

```python
#detect Peaks
peaks, _ = scs.find_peaks(y_int, height=8, distance=1000)

#plot
plt.plot(x_int, y_int)
plt.scatter(x, y)
plt.plot(x_int[peaks], y_int[peaks], 'rx')
plt.xlabel('lambda')
plt.ylabel('Csca')
plt.show()
```
